```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_players = pd.read_csv('players.csv')
df_valuations = pd.read_csv('player_valuations.csv')
df_transfers = pd.read_csv('transfers.csv')
df_appearances = pd.read_csv('appearances.csv')
df_clubs = pd.read_csv('clubs.csv')

df_players['date_of_birth'] = pd.to_datetime(df_players['date_of_birth'])
df_valuations['date'] = pd.to_datetime(df_valuations['date'])
df_transfers['transfer_date'] = pd.to_datetime(df_transfers['transfer_date'])
```

```python
df_transfers_paid = df_transfers[df_transfers['transfer_fee'].notna()]
```

```python
df_transfers_paid = df_transfers_paid[df_transfers_paid['transfer_fee'] > 0].copy()
df_transfers_paid = df_transfers_paid[df_transfers_paid['market_value_in_eur'].notna()]
```

```python
df_transfers_paid
```

```python
df_transfers_paid['net_difference'] = df_transfers_paid['transfer_fee'] - df_transfers_paid['market_value_in_eur']
```

```python
df_merged = df_transfers_paid.merge(df_players, how='left', on='player_id')
df_merged.info()
```

```python
df_merged['age_at_transfer'] = df_merged['transfer_date'].dt.year - df_merged['date_of_birth'].dt.year - \
    ((df_merged['transfer_date'].dt.month < df_merged['date_of_birth'].dt.month) | \
    ((df_merged['transfer_date'].dt.month == df_merged['date_of_birth'].dt.month) & \
    (df_merged['transfer_date'].dt.day < df_merged['date_of_birth'].dt.day)))

df_merged
```

```python
top_n = 20
top_nationalities = df_merged['country_of_citizenship'].value_counts().head(top_n).index.tolist()
print(top_nationalities)
```

```python
df_top_nationalities = df_merged[df_merged['country_of_citizenship'].isin(top_nationalities)].copy()
```

```python
mean_median_diff = df_top_nationalities.groupby('country_of_citizenship')['net_difference'].agg(['mean', 'median'])
```

```python
mean_median_diff = mean_median_diff.sort_values(by='median', ascending=False)

# 4. Create the Grouped Bar Plot
plt.figure(figsize=(14, 7)) # Adjust figure size as needed

# Create the bar plot
mean_median_diff.plot(kind='bar', figsize=(14, 7), rot=45)

plt.title('Mean and Median Net Difference by Top Nationalities')
plt.ylabel('Net Difference (€)')
plt.xlabel('Country of Citizenship')
plt.legend(title='Statistic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()

print("\nMean and Median Net Difference by Top Nationalities:")
print(mean_median_diff)
```

```python
percentile_05_by_country = df_top_nationalities.groupby('country_of_citizenship')['net_difference'].transform(lambda x: x.quantile(0.05))
percentile_95_by_country = df_top_nationalities.groupby('country_of_citizenship')['net_difference'].transform(lambda x: x.quantile(0.95))

df_filtered_outliers = df_top_nationalities[
    (df_top_nationalities['net_difference'] >= percentile_05_by_country) &
    (df_top_nationalities['net_difference'] <= percentile_95_by_country)
].copy()

mean_median_filtered_diff = df_filtered_outliers.groupby('country_of_citizenship')['net_difference'].agg(['mean', 'median'])

# Sort by median for better visualization
mean_median_filtered_diff = mean_median_filtered_diff.sort_values(by='median', ascending=False)

# 4. Plot the results to see the impact
plt.figure(figsize=(14, 7))
mean_median_filtered_diff.plot(kind='bar', figsize=(14, 7), rot=45)
plt.title(f'Mean and Median Net Difference by Top Nationalities (Top 5% Outliers Removed)')
plt.ylabel('Net Difference (€)')
plt.xlabel('Country of Citizenship')
plt.legend(title='Statistic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\nMean and Median Net Difference by Top Nationalities (Top 5% Outliers Removed):")
print(mean_median_filtered_diff)
```
